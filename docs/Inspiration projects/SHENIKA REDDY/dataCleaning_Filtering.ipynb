```python
import pandas as pd
import numpy as np

df = pd.read_csv('./283_2_FoodWebDataBase_2018_12_10.csv')

df_clean = df.copy()

# change object columns to category
df_clean[df_clean.select_dtypes(['object']).columns] = df.select_dtypes(['object']).apply(lambda x: x.astype('category'))
display(df_clean.info())
```

```python
# only include species
df_clean.loc[df_clean['con.taxonomy.level'] == 'species']
df_clean.loc[df_clean['res.taxonomy.level'] == 'species']

df_clean = df_clean[['autoID',
                     'interaction.type',
                     'con.taxonomy',
                     'con.mass.mean.g.',
                     'res.taxonomy',
                     'res.mass.mean.g.',
                     'geographic.location',
                     'longitude',
                     'latitude',
                     'ecosystem.type',
                     'foodweb.name']]
df_clean.to_csv('/content/FoodWebDataBase.csv', index=False)
```

```python
display(df_clean.sample(n=10))
```

```python
# replace -999.0 with null for easier counting
df_clean = df_clean.replace(-999.0,np.NaN)

for col in df_clean.columns:
    print('\t%s: %d' % (col,df_clean[col].isna().sum()))
```

```python
# observe different metabolic types
print(df_clean['con.metabolic.type'].unique())
```

```python
# Filter for Ythan Estuary, Scotland marine food web for testing
mask = df_clean['longitude'].astype(str).str.contains('-2.*')

df_filtered = df_clean[mask]

df_reduced = df_filtered[['autoID', 'con.taxonomy', 'res.taxonomy', 'interaction.type', 'foodweb.name']]

df_reduced.to_csv('/content/ScotMarine_FoodWeb.csv', index=False)
```

## **These are the 5 food webs filtered from Portugal**

```python
# L1P1 Portugal Marine
mask = df_clean['longitude'].astype(str).str.contains('-9.340457')

df_filtered = df_clean[mask]

df_reduced = df_filtered[['autoID', 'interaction.type', 'con.taxonomy', 'con.mass.mean.g.', 'res.taxonomy', 'res.mass.mean.g.', 'foodweb.name']]

df_reduced['autoID'] = df_reduced.groupby(['con.taxonomy', 'res.taxonomy']).ngroup()

df_reduced.to_csv('/content/Portugal_L1P1.csv', index=False)
display(df_reduced.sample(5))
```
